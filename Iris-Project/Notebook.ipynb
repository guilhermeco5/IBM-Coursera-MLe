# Project of iris classification - Utilizing from my IBM annotations    

## Installing packages

```python
from sklearn.datasets import load_iris
import pandas as pd

# Carrega o dataset
iris = load_iris()

# Converte para DataFrame (mais fácil de trabalhar)
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]
```
